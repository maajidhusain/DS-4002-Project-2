# Exploratory Data Analysis

## Create functions for the following EDA purposes
- Number of missing values
- Line plot of the data
- ACF series plot to examine lags
    - Correlation values outside of the confidence intervals indivate likely correlation and not just statistical fluke
- PACF series plot to examine lags

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf 
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
def missing_val_counts(df):
    return df.isnull().sum()
```

```python
def ts_plot(df, col_name, ticker_name):
    plt.figure(figsize=(10, 6))
    plt.plot(df.Date, df[col_name], label=col_name)
    plt.xlabel("Date")
    plt.ylabel(col_name)
    plt.title(f"{ticker_name} and {col_name} over time")
    plt.legend()
    plt.show()
```

```python
def print_end():
    print("================================================")
    print("===================== DONE =====================")
    print("================================================")
    print("\n\n")
```

```python
companies = {
    "NVIDIA": "NVDA",
    "Microsoft": "MSFT",
    "Meta Platforms": "META",
    "Apple": "AAPL",
    "Alphabet_Google": "GOOGL"  # <- want to focus on google
}

for company, ticker in companies.items():
    df = pd.read_csv(f"../DATA/{company}.csv")

    print(f"Missing Values for {company} ({ticker}):")
    print(missing_val_counts(df))

    print("\n")

    ts_plot(df, "Close", company)


    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    plot_acf(df["Close"], lags=20, title=f"{ticker} Close ACF Plot", ax=ax[0])
    plot_pacf(df["Close"], lags=20, title=f"{ticker} Close PACF Plot", ax=ax[1])
    plt.show()
    
    print_end()
```
